## What is Machine Learning

Machine Learning is the semi automated extraction of knowledge from data.

Machine learning is not an automated process as it requires thoughtful decissions to be made by a human

## Main categories of ML

* **Supervised learning**. Makes predictions using data. In this type of ML there is a specific outcome that we are trying to predict (e.g. spam/not spam, income)

* **Unsupervised learning**. Extract structure from the data. There is no "right answer" (e.g. clustering some data. There is no right answer on how many clusters there are, which cluster each individual belongs to or even how to describe each cluster).



## How does it work?

The process of supervised machine learning can be broadly described by two steps:

* **Train** a machine learning algorithm with labeled data, that is data that consists on attributes (features) and whose outcome is already known.

* **Predict** the outcome of new unseen data, for which the label is unknown, using the algorithm you already trained.

The expected result from a ML model is that the final model *generalizes* in the sense that it accurately predicts the future outcomes.
